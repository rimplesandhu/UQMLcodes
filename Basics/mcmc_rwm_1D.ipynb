```python
#!/usr/bin/python
import numpy as np
import numpy.linalg as la
import scipy.stats as st
import matplotlib.pyplot as plt
import matplotlib as mpl
import os, math, sys
```

```python
def posterior(x):
    return np.exp(-(x-1)**2/1**2)
```

```python
Ns = 1000
d =1
Xspls = [0]*Ns
Xspls[0] = 15
SigProp = 2
for i in range(0,Ns-2):
    Xprop = Xspls[i] + np.random.normal(0,SigProp)
    if posterior(Xprop)/posterior(Xspls[i]) > np.random.uniform(0,1):
        Xspls[i+1] = Xprop
    else:
        Xspls[i+1] = Xspls[i]
```

```python
fig = plt.figure(1)
plt.plot(Xspls,'r.')
plt.xlabel('MCMC samples')
```
