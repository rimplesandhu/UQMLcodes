## Generate noisy observations

###### Assumptions: polynomial basis functions, scalar output, scalar input.
###### Author: Rimple Sandhu, Carleton University; Email: rimple_sandhu@outlook.com

```python
#!/usr/bin/python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os, math, sys
np.random.seed(88)
os.system('rm *.pdf *.eps *.png');
```

#### Problem definition

```python
d = 1 # dimension of input X
M = 19 # number of observations [50] 20: case with unimodal evid
rho = 500 # precision of measurement noise [25] 1
Tmax = 20.0
Wamp = [1,0]
Wfreq = [5,2]
mpl.rcParams.update({'font.size':12})
```

#### Generate noisy observations

```python
def fn(t):
    return Wamp[0]*np.sin(Wfreq[0]*t) + Wamp[1]*np.sin(Wfreq[1]*t)
X = np.linspace(0,Tmax,M)
Y = [0]*M
for j in range(0,M):
    Y[j] = fn(X[j]) + np.random.randn()/np.sqrt(rho)
np.savetxt('Ydata.dat',Y)
np.savetxt('Xdata.dat',X)
```

#### Plot noisy vs true observations

```python
Mtrue = 500;
Ytrue = [0]*Mtrue
Xtrue = np.linspace(0,Tmax,Mtrue)
for j in range(0,Mtrue):
    Ytrue[j] = fn(Xtrue[j])
fig = plt.figure(1)
plt.plot(X,Y,'r*',label='Observations')
plt.plot(Xtrue,-np.sin(0.65*Xtrue),'-r')
plt.plot(Xtrue,Ytrue,'b-',label='Truth')
plt.grid(True)
plt.ylabel('Y')
plt.xlabel('X')
#plt.legend(loc='best', numpoints = 1)
plt.savefig('0_observations.eps',bbox_inches='tight')
```
