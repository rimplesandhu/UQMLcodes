```python
#!/usr/bin/python
import os, math, sys, random
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
polyord=[2,5,7]
fsbln = [7,8,11]
fsblf = [4,20,25]
fsbl = [4,25,50]
```

```python
fig = plt.figure(1)
plt.plot(polyord,fsbl,'r*')
plt.plot(polyord,fsbl,'r-',label='SBL')
plt.plot(polyord,fsblf,'b*')
plt.plot(polyord,fsblf,'b-',label='Fast SBL')
plt.plot(polyord,fsbln,'k*')
plt.plot(polyord,fsbln,'k-',label='NSBL (Proposed)')
plt.grid(True)
plt.ylabel('Number of evidence computations')
plt.xlabel('dimension (number of basis or hyperparameters)')
plt.legend(loc='best', numpoints = 1)
plt.savefig('compare_sbl.eps',bbox_inches='tight')
```
