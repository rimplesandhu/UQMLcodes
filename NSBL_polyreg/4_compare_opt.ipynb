```python
#!/usr/bin/python
import os, math, sys, random
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
Fsblo = np.loadtxt('SBLorig_feval.dat')
Fsblf = np.loadtxt('SBLfast_feval.dat')
Fsbln = np.loadtxt('SBLnl_feval.dat')
```

```python
fig = plt.figure(1)
plt.plot(Fsblo,'r*')
plt.plot(Fsblo,'r-',label='SBL')
plt.plot(Fsblf,'b*')
plt.plot(Fsblf,'b-',label='Fast SBL')
plt.plot(Fsbln,'k*')
plt.plot(Fsbln,'k-',label='NSBL (Proposed)')
plt.grid(True)
plt.xlim([0,40])
plt.ylabel('Log-evidence')
plt.xlabel('Evidence computation count')
plt.legend(loc='best', numpoints = 1)
plt.savefig('compare_sbl.eps',bbox_inches='tight')
```
